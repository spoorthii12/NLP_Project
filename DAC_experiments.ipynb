```python
import torch
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification

# Load the trained model and tokenizer
tokenizer = DistilBertTokenizer.from_pretrained('./dialogue_model')
model = DistilBertForSequenceClassification.from_pretrained('./dialogue_model')

# Mapping labels back to names
dialogue_act_labels = {1: "inform", 2: "question", 3: "directive", 4: "commissive"}
emotion_labels = {0: "no emotion", 1: "anger", 2: "disgust", 3: "fear", 4: "happiness", 5: "sadness", 6: "surprise"}

def predict(sentence):
    # Tokenize input sentence
    inputs = tokenizer(sentence, padding='max_length', truncation=True, max_length=128, return_tensors='pt')

    # Perform inference
    with torch.no_grad():
        outputs = model(input_ids=inputs['input_ids'], attention_mask=inputs['attention_mask'])

    # Extract predictions
    act_pred = torch.argmax(outputs.logits[0]).item()  # Adjust for dialogue act logits
    emotion_pred = torch.argmax(outputs.logits[1]).item()  # Adjust for emotion logits

    # Decode predictions
    act = dialogue_act_labels[act_pred]
    emotion = emotion_labels[emotion_pred]

    return act, emotion

# CLI interface
if __name__ == "__main__":
    print("Welcome to Dialogue Act and Emotion Classifier!")
    while True:
        sentence = input("Enter a sentence (or type 'exit' to quit): ")
        if sentence.lower() == 'exit':
            print("Goodbye!")
            break
        act, emotion = predict(sentence)
        print(f"Predicted Dialogue Act: {act}")
        print(f"Predicted Emotion: {emotion}")
```

```python
from transformers import DistilBertTokenizer, DistilBertModel, Trainer, TrainingArguments, EarlyStoppingCallback
from transformers.modeling_outputs import SequenceClassifierOutput
import torch
import pandas as pd
import os
import ast
from collections import Counter


# Disable W&B syncing for offline usage
os.environ["WANDB_MODE"] = "offline"

# Load the tokenizer
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")


# Function for basic text preprocessing
def preprocess_text(text):
    return " ".join(text.lower().strip().split())  # Simplified preprocessing


# Function to load and preprocess the dataset
def load_and_preprocess(filepath):
    df = pd.read_csv(filepath)
    df['dialog'] = df['dialog'].apply(preprocess_text)

    # Fix improperly formatted lists in the 'act' and 'emotion' columns
    if 'act' in df.columns:
        df['act'] = df['act'].apply(
            lambda x: str(x).replace(" ", ",") if isinstance(x, str) and "[" in x else x
        )
        df['act'] = df['act'].apply(
            lambda x: ast.literal_eval(x) if isinstance(x, str) and "[" in x else x
        )
        df['act'] = df['act'].apply(
            lambda x: max(set(x), key=x.count) if isinstance(x, list) else x
        )

    if 'emotion' in df.columns:
        df['emotion'] = df['emotion'].apply(
            lambda x: str(x).replace(" ", ",") if isinstance(x, str) and "[" in x else x
        )
        df['emotion'] = df['emotion'].apply(
            lambda x: ast.literal_eval(x) if isinstance(x, str) and "[" in x else x
        )
        df['emotion'] = df['emotion'].apply(
            lambda x: max(set(x), key=x.count) if isinstance(x, list) else x
        )

    return df


# Tokenization function
def tokenize_data(df):
    tokenized = tokenizer(
        df['dialog'].tolist(),
        padding='max_length',
        truncation=True,
        max_length=128,
        return_tensors='pt'
    )
    df['input_ids'] = tokenized['input_ids'].tolist()
    df['attention_mask'] = tokenized['attention_mask'].tolist()
    return df


# Prepare data for PyTorch Dataset
def prepare_dataset(df):
    return df[['input_ids', 'attention_mask']].values.tolist(), df[['act', 'emotion']].values.tolist()


# Define PyTorch Dataset
class DialogueDataset(torch.utils.data.Dataset):
    def __init__(self, inputs, labels):
        self.inputs = inputs
        self.labels = labels

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        item = {
            'input_ids': torch.tensor(self.inputs[idx][0], dtype=torch.long),
            'attention_mask': torch.tensor(self.inputs[idx][1], dtype=torch.long),
            'act_label': torch.tensor(self.labels[idx][0], dtype=torch.long),
            'emotion_label': torch.tensor(self.labels[idx][1], dtype=torch.long),
        }
        return item


# Define Multi-Task Model with Weighted Loss & Improved Emotion Classification
class MultiTaskDistilBERT(torch.nn.Module):
    def __init__(self, act_weights=None, emotion_weights=None):
        """
        Multi-task DistilBERT model with enhanced emotion classification.
        :param act_weights: Class weights for dialogue act loss
        :param emotion_weights: Class weights for emotion loss
        """
        super().__init__()
        self.bert = DistilBertModel.from_pretrained("distilbert-base-uncased", output_hidden_states=True)

        # Fully connected layers for multi-task learning
        self.shared_dropout = torch.nn.Dropout(0.3)  # Introduced regularization

        # Multi-task heads
        self.act_classifier = torch.nn.Sequential(
            torch.nn.Linear(self.bert.config.hidden_size, 128),
            torch.nn.GELU(),
            torch.nn.Dropout(0.3),
            torch.nn.Linear(128, 5)  # 5 classes for acts
        )

        self.emotion_fc_1 = torch.nn.Sequential(
            torch.nn.Linear(self.bert.config.hidden_size, 128),
            torch.nn.GELU(),
            torch.nn.Dropout(0.3)
        )

        self.emotion_fc_2 = torch.nn.Linear(128, 7)  # Final emotion classifier layer

        # Loss functions
        self.act_loss_fn = torch.nn.CrossEntropyLoss()
        self.emotion_loss_fn = torch.nn.CrossEntropyLoss(weight=torch.tensor(emotion_weights, dtype=torch.float)) if emotion_weights is not None else torch.nn.CrossEntropyLoss()

    def forward(self, input_ids, attention_mask, act_label=None, emotion_label=None):
        # Pass through DistilBERT
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        hidden_states = outputs.hidden_states[-1]
        pooled_output = hidden_states[:, 0]  # CLS token pooling

        # Shared dropout for regularization
        pooled_output = self.shared_dropout(pooled_output)

        # Dialogue Act prediction
        act_logits = self.act_classifier(pooled_output)

        # Emotion classification with non-linear transformation
        emotion_features = self.emotion_fc_1(pooled_output)
        emotion_logits = self.emotion_fc_2(emotion_features)

        # Compute losses if provided
        loss = None
        if act_label is not None and emotion_label is not None:
            # Calculate the dialogue act and emotion loss
            act_loss = self.act_loss_fn(act_logits, act_label)
            emotion_loss = self.emotion_loss_fn(emotion_logits, emotion_label)
            loss = act_loss + emotion_loss

        return {
            "loss": loss,
            "act_logits": act_logits,
            "emotion_logits": emotion_logits
        }


# Load datasets
train_df = load_and_preprocess("/content/drive/MyDrive/NLP-project/Dataset/train.csv")
test_df = load_and_preprocess("/content/drive/MyDrive/NLP-project/Dataset/test.csv")
val_df = load_and_preprocess("/content/drive/MyDrive/NLP-project/Dataset/validation.csv")

# Tokenize datasets
train_df = tokenize_data(train_df)
test_df = tokenize_data(test_df)
val_df = tokenize_data(val_df)

# Prepare data
X_train, y_train = prepare_dataset(train_df)
X_test, y_test = prepare_dataset(test_df)
X_val, y_val = prepare_dataset(val_df)

# Compute emotion class weights dynamically to handle imbalance
emotion_counts = Counter([label[1] for label in y_train])
total_count = sum(emotion_counts.values())
emotion_weights = [total_count / emotion_counts.get(i, 1) for i in range(7)]

# Create datasets
train_dataset = DialogueDataset(X_train, y_train)
val_dataset = DialogueDataset(X_val, y_val)

# Define model with computed emotion weights
model = MultiTaskDistilBERT(emotion_weights=emotion_weights)

# Training arguments
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    logging_dir='./logs',
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    learning_rate=5e-5,  # Set learning rate to a lower value
    warmup_steps=50  # Gradually ramp up learning rate
)



from transformers import EarlyStoppingCallback

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)],  # Implement early stopping here
)


# Train the model
trainer.train()

# Save the model & tokenizer
model_path = "./dialogue_model"
trainer.save_model(model_path)
tokenizer.save_pretrained(model_path)

print(f"Model saved to: {model_path}")
```

```python
from transformers import DistilBertTokenizer, DistilBertModel
import torch
import os
from torch.nn import Module
from transformers.modeling_outputs import SequenceClassifierOutput

# Ensure WANDB is offline
os.environ["WANDB_MODE"] = "offline"

# Define mappings for dialogue acts and emotions
dialogue_acts = {0: "__dummy__", 1: "inform", 2: "question", 3: "directive", 4: "commissive"}
emotions = {0: "no emotion", 1: "anger", 2: "disgust", 3: "fear", 4: "happiness", 5: "sadness", 6: "surprise"}

# Define Multi-Task Model Architecture (same as training)
class MultiTaskDistilBERT(Module):
    def __init__(self):
        super().__init__()
        # Load DistilBERT Model
        self.bert = DistilBertModel.from_pretrained("distilbert-base-uncased", output_hidden_states=True)

        # Linear layers for classification
        self.act_classifier = torch.nn.Linear(self.bert.config.hidden_size, 5)  # 5 classes for acts
        self.emotion_classifier = torch.nn.Linear(self.bert.config.hidden_size, 7)  # 7 classes for emotions

    def forward(self, input_ids, attention_mask, act_label=None, emotion_label=None):
        # Pass inputs through DistilBERT
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        hidden_states = outputs.hidden_states[-1]  # Last hidden state from the model
        pooled_output = hidden_states[:, 0]  # [CLS] token representation

        # Perform classification
        act_logits = self.act_classifier(pooled_output)
        emotion_logits = self.emotion_classifier(pooled_output)

        # Calculate loss if labels are provided
        loss = None
        if act_label is not None and emotion_label is not None:
            act_loss_fn = torch.nn.CrossEntropyLoss()
            emotion_loss_fn = torch.nn.CrossEntropyLoss()
            act_loss = act_loss_fn(act_logits, act_label)
            emotion_loss = emotion_loss_fn(emotion_logits, emotion_label)

            # Combine losses
            loss = act_loss + emotion_loss

        # Return the logits and loss
        return SequenceClassifierOutput(
            loss=loss,
            logits=(act_logits, emotion_logits)
        )


def load_model_and_tokenizer(model_path="./dialogue_model"):
    tokenizer = DistilBertTokenizer.from_pretrained(model_path)
    model = MultiTaskDistilBERT()

    # Load model weights from safetensors
    from safetensors.torch import load_file
    state_dict = load_file(f"{model_path}/model.safetensors")
    model.load_state_dict(state_dict, strict=False)  # Allow missing/extra keys

    model.eval()  # Set model to evaluation mode
    return model, tokenizer



# Prediction function
def predict(text, model, tokenizer):
    # Preprocess text
    text = " ".join(text.lower().strip().split())

    # Tokenize input
    inputs = tokenizer(text, truncation=True, padding='max_length', max_length=128, return_tensors="pt")

    # Make predictions
    with torch.no_grad():
        outputs = model(
            input_ids=inputs['input_ids'],
            attention_mask=inputs['attention_mask']
        )
        act_logits, emotion_logits = outputs.logits
        act_pred = torch.argmax(act_logits, dim=1).item()
        emotion_pred = torch.argmax(emotion_logits, dim=1).item()

    # Decode predictions
    return f"Dialogue Act: {dialogue_acts[act_pred]}, Emotion: {emotions[emotion_pred]}"


# Interactive CLI
def interactive_cli():
    print("Interactive Dialogue Act & Emotion Classifier")
    print("Type 'exit' to quit.\n")

    # Load model and tokenizer
    model_path = "./dialogue_model"
    model, tokenizer = load_model_and_tokenizer(model_path)

    while True:
        user_input = input("Enter a dialog: ")
        if user_input.lower() == "exit":
            print("Exiting... Goodbye!")
            break
        result = predict(user_input, model, tokenizer)
        print(result)


# Run CLI
if __name__ == "__main__":
    interactive_cli()
```

```python
import os
import torch
import pandas as pd
import numpy as np
import nlpaug.augmenter.word as naw
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import optuna
import nltk

# Disable W&B syncing for offline usage
os.environ["WANDB_MODE"] = "offline"

# Download necessary NLTK resources only once
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)

# Load the tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Custom stopword list
STOPWORDS = {
    "i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours",
    "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself",
    "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which",
    "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be",
    "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an",
    "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by",
    "for", "with", "about", "against", "between", "into", "through", "during", "before",
    "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over",
    "under", "again", "further", "then", "once", "here", "there", "when", "where", "why",
    "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such",
    "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can",
    "will", "just", "don", "should", "now"
}

# Function for removing stopwords
def remove_stopwords(text):
    words = text.split()  # Basic tokenization using split()
    return " ".join([word for word in words if word.lower() not in STOPWORDS])

# Function for text augmentation
def augment_text(text):
    augmenter = naw.SynonymAug(aug_src='wordnet')
    try:
        augmented_text = augmenter.augment(text)
        return augmented_text if augmented_text else text
    except Exception as e:
        return text  # Fallback to original text if augmentation fails

# Function for advanced preprocessing
def preprocess_text(text):
    text = text.lower().strip()
    text = remove_stopwords(text)
    return augment_text(text)

# Load datasets
train_df = pd.read_csv("/content/train.csv")
test_df = pd.read_csv("/content/test.csv")
val_df = pd.read_csv("/content/validation.csv")

# Preprocess dialog column
for df in [train_df, test_df, val_df]:
    df['dialog'] = df['dialog'].apply(preprocess_text)

# Tokenization function
def tokenize_data(df):
    tokenized = tokenizer(
        df['dialog'].tolist(),
        padding='max_length',
        truncation=True,
        max_length=128,
        return_tensors='pt'
    )
    df['input_ids'] = tokenized['input_ids'].tolist()
    df['attention_mask'] = tokenized['attention_mask'].tolist()
    return df

# Tokenize the datasets
train_df = tokenize_data(train_df)
val_df = tokenize_data(val_df)

# Prepare data for PyTorch Dataset
def prepare_dataset(df):
    return df[['input_ids', 'attention_mask']].values.tolist(), df['act'].tolist()

X_train, y_train = prepare_dataset(train_df)
X_val, y_val = prepare_dataset(val_df)

# Define PyTorch Dataset
import ast  # For parsing corrected strings

def fix_malformed_list(x):
    if isinstance(x, str) and "[" in x and "]" in x:
        return x.replace(" ", ",")  # Fix invalid list formats by adding commas
    return x

class DialogueDataset(torch.utils.data.Dataset):
    def __init__(self, inputs, labels):
        self.inputs = inputs
        # Fix malformed lists and parse them
        self.labels = []
        for label in labels:
            fixed_label = fix_malformed_list(label)
            if isinstance(fixed_label, str) and "[" in fixed_label and "]" in fixed_label:
                # Parse the corrected list
                parsed_list = ast.literal_eval(fixed_label)
                # Pad the list to a fixed length
                padded_list = parsed_list + [0] * (10 - len(parsed_list))  # Pad with 0s up to length 10
                self.labels.append(padded_list[:10])  # Truncate to length 10 if longer
            else:
                # Pad single integer labels to the fixed length
                self.labels.append([int(label)] + [0] * 9)  # Pad with 0s up to length 10

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        return {
            'input_ids': torch.tensor(self.inputs[idx][0], dtype=torch.long),
            'attention_mask': torch.tensor(self.inputs[idx][1], dtype=torch.long),
            'labels': torch.tensor(self.labels[idx], dtype=torch.long),
        }


# Create datasets
train_dataset = DialogueDataset(X_train, y_train)
val_dataset = DialogueDataset(X_val, y_val)

# Load the model
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=5)

# Define the evaluation metrics
def compute_metrics(pred):
    labels = pred.label_ids
    preds = np.argmax(pred.predictions, axis=1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average="weighted")
    cm = confusion_matrix(labels, preds)
    return {
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "confusion_matrix": cm.tolist()  # Convert to list for JSON compatibility
    }

# Define objective function for Optuna
def objective(trial):
    # Hyperparameters to optimize
    batch_size = trial.suggest_categorical("batch_size", [8, 16, 32])
    learning_rate = trial.suggest_loguniform("learning_rate", 1e-5, 5e-5)
    weight_decay = trial.suggest_loguniform("weight_decay", 1e-6, 1e-2)

    # Training arguments
    training_args = TrainingArguments(
        output_dir='./results',
        num_train_epochs=3,
        per_device_train_batch_size=batch_size,
        evaluation_strategy="epoch",
        save_strategy="epoch",
        logging_dir='./logs',
        load_best_model_at_end=True,
        metric_for_best_model="eval_f1",
        learning_rate=learning_rate,
        weight_decay=weight_decay,
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        compute_metrics=compute_metrics,
    )

    trainer.train()
    eval_results = trainer.evaluate(eval_dataset=val_dataset)
    return eval_results["eval_f1"]

# Run Optuna for hyperparameter tuning
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=10)

# Best hyperparameters
print(f"Best hyperparameters: {study.best_params}")

# Train with best hyperparameters
best_params = study.best_params
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=5,
    per_device_train_batch_size=best_params["batch_size"],
    evaluation_strategy="epoch",
    save_strategy="epoch",
    logging_dir='./logs',
    load_best_model_at_end=True,
    metric_for_best_model="eval_f1",
    learning_rate=best_params["learning_rate"],
    weight_decay=best_params["weight_decay"],
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)

# Train the final model
trainer.train()

# Evaluate the model
evaluation_results = trainer.evaluate(eval_dataset=val_dataset)
print(f"Evaluation Results: {evaluation_results}")

# Save model and tokenizer
model_path = "./dialogue_model"
trainer.save_model(model_path)
tokenizer.save_pretrained(model_path)

print(f"Model saved to: {model_path}")

```
